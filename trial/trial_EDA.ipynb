## 概形チェック

```python
def stats(df, digits=2):
    cols = df.columns
    
    max_cols = []
    min_cols = []
    for i in cols:
        max_cols.append(df[i].value_counts().max())
        min_cols.append(df[i].value_counts().min())

    l_df = len(df)
    n_df = df.isna().sum()
    
    result = pd.DataFrame({
        'name': cols,
        'dtype': df.dtypes,
        'len': l_df,
        'nulls': n_df,
        'null percent': round(n_df / l_df * 100, digits),
        'nunique': df.nunique(),
        'value_counts max': max_cols,
        'value_counts min': min_cols
    }).reset_index(drop=True)
    
    return result
```

```python
stats(df_train)
```

## ヒストグラム

### コード

```python
def histplot(df, col, interval=10, color="cyan", edgecolor="black", fontsize=16):
    bin_min = int(df[col].min())
    bin_max = int(df[col].max())
    edges = range(bin_min, bin_max, interval)
    plt.hist(df[col], bins=edges, histtype="bar", color=color, edgecolor=edgecolor)

    plt.title("histgram\ncolumn: " + col, fontsize=fontsize)
    plt.show()
```

### 可視化

```python
histplot(df_train, "Age", interval=5)
```

```python
histplot(df_train, "Fare", interval=20)
```

## 棒グラフ

### コード

```python
def barplot(df, col, type="bar", color="", edgecolor="black", width=0.7, fontsize=16):
    if color == "":
        if type == "bar":
            color = "cyan"
        elif type == "cat":
            color = "orange"

    target = df[col].value_counts().sort_index(ascending=True, inplace=False)
    if type == "cat":
        num = len(df[col])
        target = target / num

    target.plot(kind="bar", color=color, edgecolor=edgecolor, width=width, xlabel="")

    plt.title("barplot\ncolumn: " + col, fontsize=fontsize)
    plt.show()
```

### 可視化

```python
barplot(df_train, "Survived")
```

### コード

```python
def target_barplot(df, target, col, color=["cyan", "orange"], edgecolor="black", width=0.7, fontsize=16):
    obj_y = np.sort(df[target].unique())

    target0 = df.loc[(df[target] == obj_y[0])][col].value_counts()
    target1 = df.loc[(df[target] == obj_y[1])][col].value_counts()
    result = pd.DataFrame({
        obj_str[0]: target0,
        obj_str[1]: target1
    })

    result.plot(kind="bar", color=color, edgecolor=edgecolor, width=width, xlabel="")
    plt.title("target barplot\ntarget: " + target + ", column: " + col, fontsize=fontsize)
    plt.show()
```

```python
target_barplot(df_train, "Survived", "Sex")
```

```python
target_barplot(df_train, "Survived", "Parch")
```

```python
target_barplot(df_train, "Survived", "SibSp")
```

### コード

```python
def target_histplot(df, target, col, bins=50, color=["cyan", "orange"], edgecolor=["black", "black"], linewidth=2, fontsize=16):
    obj_y = np.sort(df[target].unique())

    df[df[target]==obj_y[0]][col].plot(kind="hist", bins=bins, alpha=0.2, label=str(obj_y[0]), edgecolor=edgecolor[0], linewidth=linewidth)
    df[df[target]==obj_y[1]][col].plot(kind="hist", bins=bins, alpha=0.2, label=str(obj_y[1]), edgecolor=edgecolor[1], linewidth=linewidth)

    plt.title("target histplot\ntarget: " + target + ", column: " + col, fontsize=fontsize)
    plt.legend()
    plt.ylabel("")
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
target_histplot(df, "Survived", "Age")
```
